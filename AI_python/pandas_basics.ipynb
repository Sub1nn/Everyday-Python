### Handling Missing Data (NaN)

Handling NaN (Not a Number) values is important because they can interfere with data analysis and machine learning models

 **Statistical Analysis**: calculating the mean, median, variance, or correlation, will return incorrect results or fail if there are NaN values in the data
 
 `data = pd.Series([1, 2, np.nan, 4, 5])`
 
 `print(data.mean())`
 
 `print(data.mean(skipna=True))`
 
 **Problems with Machine Learning Models**: Algorithms like linear regression, decision trees, or neural networks require complete data. If NaN values are present, they either need to be removed or replaced with something meaningful to ensure proper model training.

```python
import numpy as np
```

```python
data = pd.Series([1, 2, np.nan, 4, 5])
data.mean(skipna=True)
```

# Data Frame

- Pandas DataFrame is two-dimensional size-mutable, potentially heterogeneous tabular data structure with labeled axes (rows and columns). 

- A Data frame is a two-dimensional data structure, i.e., data is aligned in a tabular fashion in rows and columns. Pandas DataFrame consists of three principal components, the data, rows, and columns.


Pandas DataFrame is created by loading the datasets from existing storage (which can be a SQL database, a CSV file, or an Excel file).



### Creating Data Frame

```python
import pandas as pd
data={
    "name":["Bill","Tom","Tim","John","Alex","Vanessa","Kate"],      
    "score":[90,80,85,75,95,60,65],      
    "sport":["Wrestling","Football","Skiing","Swimming","Tennis","Karete","Surfing"],      
    "gender":["M","M","M","M","F","F","F"]
}

df=pd.DataFrame(data)
```

### Accessing and indexing Data in Data Frame

```python
df
```

```python
# Accessing a single column
name_data_frame = df["name"]

# # Accessing multiple columns
df[["name","sport", "score"]]
```

```python
# # Accessing a specific row by index
print(df.iloc[2])  # Third row (index 2)

# # Accessing a range of rows using slicing
print(df.iloc[1:4, 0:2])  # Rows with indices 1, 2, 3
```

Indexing with `.loc[]` and `.iloc[]`

- `.loc[]`: Label-based indexing. It works with labels (row/column names).
- `.iloc[]`: Position-based indexing. It works with integer positions (row/column indices).


```python
df.set_index("name", inplace=True)
```

```python
df
```

```python
df.reset_index(inplace=True)
df
```

```python
df
```

```python
# Accessing row by label (name)
print(df.loc["Bill"])  # Row with index 1
```

```python

# Accessing specific value using labels
print(df.loc["Bill", "score"])  # score of the person at index 1
```

```python
# Accessing row by integer position
print(df.iloc[1])  # Second row (index 1)
```

```python
# Accessing specific value using integer positions
print(df.iloc[1, 2])  # Element at row 1, column 1
```

```python
print(df.iloc[1, [0,2]])  # Element at row 1, column 0 and 2
```

### Slicing

- With `.loc[]`: Both the start and stop are included.
- With `.iloc[]`: The stop is excluded.

```python
# loc or iloc [row slicing, column slicing]
```

```python
df
```

```python
# Using .loc for slicing (inclusive of end)
print(df.loc["Bill":"Tim", "score":"gender"])
```

```python
# Using .iloc for slicing (exclusive of end)
print(df.iloc[1:3, 0:2])
```

### Filtering Data (Boolean Indexing)

```python
# Filtering based on a condition (e.g., score > 80)
df[df["gender"] == 'M']
```

```python
# Filtering based on multiple conditions
print(df[(df["score"] > 80) & (df["gender"] == "M")])
```

### Sorting Data

```python
# Sorting by score
print(df.sort_values("score", ascending=False))
```

```python
# Sorting by multiple columns (first by sex, then by score)
print(df.sort_values(by=["gender", "score"], ascending=[True, False]))
```

### GroupBy

`groupby()` function in pandas is used to split the data into groups based on some criteria

and we can apply a function to each group independently

```python
data = {'Category': ['A', 'B', 'A', 'B', 'A'],
        'Values': [10, 20, 30, 40, 50]}

df = pd.DataFrame(data)

print(df)

# Group by 'Category' and calculate the sum of 'Values'
grouped = df.groupby('Category')

# print(grouped)

grouped.sum()
```

```python
grouped.mean()
```

## Merging, Joining, and Concatenating DataFrames

`merge()` function is used to combine two DataFrames based on a common column (like SQL joins: inner', 'left', 'right', 'outer').

python
Copy


```python
# Sample data
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [1, 2, 4], 'Age': [25, 30, 22]})

# Merge DataFrames on 'ID'
merged_df = pd.merge(df1, df2, on='ID', how='inner')  # 'inner', 'left', 'right', 'outer'
print(merged_df)
```
