# Combining & Merging Datasets in Pandas

```python
import pandas as pd
```

## Joining DataFrame

`merge()` function allows you to combine two DataFrames based on a common column (like SQL joins: 'inner', 'left', 'right', 'outer'). 

Inner Join: Only the rows with common keys (default)

```python
d1=pd.DataFrame(
    {"key":["a","b","c","c","d","e"],
     "num1":range(6)})
d2=pd.DataFrame(
    {"key":["b","c","e","f"],
     "num2":range(4)})
```

```python
print(d1)
print(d2)
```

```python
pd.merge(d1, d2)
```

```python
pd.merge(d1, d2, on='key1')
```

**Dataset having different key names**

```python
import pandas as pd

d3=pd.DataFrame(
    {"key1":["a","b","c","c","d","e"],
     "num1":range(6)})
d4=pd.DataFrame(
    {"key2":["b","c","e","f"],
     "num2":range(4)})
```

```python
print(d3)
print(d4)
```

```python
pd.merge(
    d3,d4,left_on="key1",right_on="key2"
)
```

**Outer Join**

All rows from both DataFrames, with NaN where no match is found

```python
print(d1)
print(d2)
```

```python
pd.merge(d1,d2,how="outer")
```

**Left Join**

 All rows from the left DataFrame, with matching rows from the right DataFrame (NaN for unmatched rows)

```python
pd.merge(d1,d2,how="left")
```

**Right Join**

All rows from the right DataFrame, with matching rows from the left DataFrame.

```python
pd.merge(d1,d2,how="right")
```

## Merging on index

```python
df1=pd.DataFrame(
    {"letter":["a","a","b","b","a","c"],
     "num":range(6)}) 

df2=pd.DataFrame(
    {"value":[3,5,7]},
    index=["a","b","e"])
```

```python
print(df1)
print(df2)
```

```python
pd.merge(df1,df2,
         left_on="letter",
         right_index=True)
```

```python
data1 = pd.DataFrame(
    [[1,2],[3,4],[5,6]],
    index=["a","c","d"],
    columns=["Tom","Tim"])

data2 = pd.DataFrame(
    [[7,8],[9,10],[11,12],[13,14]],
    index=["a","b","e","f"],
    columns=["Sam","Kim"])
```

```python
print(data1)
print(data2)
```

```python
pd.merge(data1,data2, 
         right_index=True, 
         left_index=True, 
         how="outer")
```

**Using `Obj.join()`**

```python
data1.merge(data2)
```

```python
data2.join(data1)
```

```python
data1.join(data2,how="outer")
```

**Multiple join**

```python
data=pd.DataFrame([[1,3],[5,7],[9,11]],            
                  index=["a","b","f"],      
                  columns=["Alex","Keta"])
data1.join([data2,data])
```

## Concatenating Along an Axis

```python
data1 = pd.Series(
    [0, 1], index=['a', 'b'])
data2 = pd.Series(
    [2,3,4], index=['c','d','e'])
data3 = pd.Series(
    [5, 6], index=['f', 'g'])
```

```python
print(data1)
print(data2)
print(data3)
```

```python
pd.concat([data1,data2,data3])
```

```python
pd.concat([data1, data2, data3], axis=1)
```

```python
data4= pd.Series([10,11,12], 
                 index=['a','b',"c"])
pd.concat([data1,data4],axis=1,join="inner")
```
