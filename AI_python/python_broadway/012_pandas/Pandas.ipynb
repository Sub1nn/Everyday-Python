### **Pandas**
Pandas is a powerful Python library for data manipulation and analysis. It provides two primary data structures:

- **Series**: A one-dimensional labeled array capable of holding any data type.
- **DataFrame**: A two-dimensional labeled data structure with columns of potentially different types.

#### Why Use Pandas?
- Easy handling of missing data.
- Tools for reading and writing data between in-memory data structures and different file formats (CSV, Excel, SQL, etc.).
- Powerful data alignment and reshaping capabilities.

---

### **Series:**

A **Series** in Pandas is a one-dimensional labeled array capable of holding any data type (e.g., integers, strings, floats, Python objects). Think of it as a column in an Excel sheet or a database table, but it can also stand alone.

---

### **Key Characteristics of a Series**
1. **Indexing**: Every element in a Series has an associated index, which is similar to row labels in a DataFrame.
2. **Homogeneous Data**: A Series can only hold one data type at a time (though it can be object type, which allows mixed types).
3. **Labeled Data**: The index provides labels for data points, making it easier to access specific values.

---

### **Creating a Series**

#### **1. From a List**

````python
import pandas as pd

data = [10, 20, 30, 40, 50]
series = pd.Series(data)
print(series)
````

#### **2. From a Scalar Value**

````python
# Create a Series from a scalar value
series = pd.Series(5, index=['a', 'b', 'c', 'd'])
print(series)
````

#### **3. From a Dictionary**

````python
data = {'a': 10, 'b': 20, 'c': 30, 'd': 40}
series = pd.Series(data)
print(series)
````

---

### **Accessing Data in a Series**

#### **1. Access by Index**

````python
# Access by position
print(series.iloc[0])  # First element
````

````python
# Access by label
print(series['a'])  # Element with label 'a'
````

#### **2. Slicing**

````python
# Access a range of elements
print(series[1:3])  # Elements from index 1 to 2
````

### **DataFrame Data Structure**

#### What is a DataFrame?
A DataFrame is like a table in a database or an Excel spreadsheet. It has rows and columns, where each column can hold a different type of data.

#### Creating a DataFrame

````python
import pandas as pd

# Creating a DataFrame from a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)
````

````python
df
````

````python
# Selecting a single column
df['Name']
````

````python
# Selecting multiple columns
df[['Name', 'City']]
````

````python
# Selecting rows by index
df.iloc[0]  # First row
````

````python
df.iloc[1:3]  # Rows 1 to 2
````

#### Filtering Data

````python
# Filter rows where Age > 25
filtered_df = df[df['Age'] > 25]
filtered_df
````

#### Adding a New Column

````python
df['Country'] = ['USA', 'USA', 'USA']
df
````

````python
# Save DataFrame to a CSV file
df.to_csv('output.csv', index=False)
````

### **Series vs DataFrame**

## **Basic Data Visualization**

### **Introduction to Matplotlib, Seaborn, and Plotly**

#### Matplotlib
Matplotlib is a low-level library for creating static, animated, and interactive plots in Python.

#### Seaborn
Seaborn is built on top of Matplotlib and provides a high-level interface for drawing attractive statistical graphics.

#### Plotly
Plotly is a library for creating interactive visualizations.

---

### **Basic Plotting Using Matplotlib**

#### Line Plot

````python
import matplotlib.pyplot as plt

# Data
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
temperatures = [22, 24, 19, 25, 21]

# Create the plot
plt.plot(days, temperatures, marker='o', color='blue', label='Temperature')

# Add title and labels
plt.title('Temperature Over a Week')
plt.xlabel('Days')
plt.ylabel('Temperature (C)')
plt.legend()

# Show the plot
plt.show()
````

````python
# Data
products = ['Mobile', 'Laptop', 'Refrigereter']
sales = [100, 150, 80]

# Create the plot
plt.bar(products, sales, color='green')

# Add title and labels
plt.title('Sales of Products')
plt.xlabel('Products')
plt.ylabel('Units Sold')

# Show the plot
plt.show()
````

### **Basic Plotting Using Seaborn**

### **Introduction to Seaborn**

Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating visually appealing and informative statistical graphics. Seaborn integrates well with Pandas, making it easy to visualize DataFrames.

---

### **Why Use Seaborn?**
1. **Simplifies Visualization**: Pre-built themes and functions reduce the effort to create plots.
2. **Built-in Statistics**: Automatically calculates and displays statistical summaries (e.g., confidence intervals).
3. **Works with DataFrames**: Directly works with Pandas DataFrames for seamless data handling.

---

### **Basic Workflow**
1. Import the library.
2. Prepare your data (preferably as a Pandas DataFrame).
3. Use Seaborn functions to create plots.
4. Customize the plots as needed.

---

### **Real-World Examples**

#### **1. Line Plot: Monthly Sales Trend**
Line plots show trends over time.

````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
        'Sales': [200, 300, 100, 500, 600]}
df = pd.DataFrame(data)

# Create the plot
sns.lineplot(x='Month', y='Sales', data=df, marker='o')

# Add title and labels
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales (in $)')
plt.show()
````

### **Difference Between Matplotlib and Seaborn**
Both **Matplotlib** and **Seaborn** are popular Python libraries for data visualization, but they have different strengths:

| Feature        | Matplotlib | Seaborn |
|---------------|-----------|---------|
| **Ease of Use** | Low (requires more code) | High (simpler syntax) |
| **Customization** | Very high | High, but built-in styles are already polished |
| **Statistical Visualization** | Limited | Built-in support for statistical plots |
| **Default Aesthetics** | Basic | Stylish and attractive |

### **Example: Matplotlib vs. Seaborn**
#### **Matplotlib Example**
- Matplotlib requires more manual configuration.
- Matplotlib works with Pandas DataFrames, but unlike Seaborn, it does not automatically recognize column names.
- We need to explicitly pass DataFrame columns to Matplotlib functions.
- Seaborn can work without a DataFrame, just like Matplotlib. 
- We can simply pass lists or arrays directly to Seaborn functions.

````python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
        'Sales': [200, 300, 100, 500, 600]}
df = pd.DataFrame(data)

# Create the line plot
plt.plot(df['Month'], df['Sales'], marker='o', linestyle='-', color='b')

# Add title and labels
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales (in $)')

# Show plot
plt.show()
````

### **Key Takeaways**
- Use **Matplotlib** for low-level, highly customizable plots.
- Use **Seaborn** when working with **dataframes** and **statistical visualizations**.
- **Seaborn** is built on top of **Matplotlib** and can be used alongside it.

## **Example: Scatter Plot in Both Matplotlib vs Seaborn**

Let's take a **real-world example** where we compare **monthly sales revenue** from two different stores over time. This example will help **visualize trends** in sales performance.

---

### ** Scenario: Sales Comparison Between Two Stores**
Imagine you are tracking sales revenue (in dollars) for **two retail stores (Store A and Store B)** over **6 months**.

---

### **1 Matplotlib Example**

````python
import matplotlib.pyplot as plt

# Real-world sales data for two stores
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
store_a_sales = [1500, 1800, 2000, 2200, 2100, 2500]
store_b_sales = [1700, 1600, 2100, 2400, 2300, 2700]

# Create line plots for both stores
plt.plot(months, store_a_sales, marker='o', linestyle='-', color='blue', label="Store A")
plt.plot(months, store_b_sales, marker='s', linestyle='--', color='red', label="Store B")

# Add title and labels
plt.title("Monthly Sales Comparison (Store A vs. Store B)")
plt.xlabel("Month")
plt.ylabel("Sales Revenue ($)")
plt.legend()  # Show legend
plt.grid(True)  # Add grid for better readability

# Show plot
plt.show()
````

---

### **2 Seaborn Example**

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Same real-world sales data
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
store_a_sales = [1500, 1800, 2000, 2200, 2100, 2500]
store_b_sales = [1700, 1600, 2100, 2400, 2300, 2700]

# Apply Seaborn style
sns.set_style("darkgrid")

# Create line plots
sns.lineplot(x=months, y=store_a_sales, marker='o', label="Store A", color='blue')
sns.lineplot(x=months, y=store_b_sales, marker='s', label="Store B", color='red', linestyle='--')

# Add title and labels
plt.title("Monthly Sales Comparison (Store A vs. Store B)")
plt.xlabel("Month")
plt.ylabel("Sales Revenue ($)")

# Show plot
plt.show()
````

### ** Real-World Insights from the Graph**
- **Which store is performing better?**   
   Store B had **higher revenue** in most months except February.  
- **Is there a trend?**   
   Both stores **show an upward trend** in sales, indicating growth.  
- **Where do stores need improvement?**   
   Store As sales **dropped slightly in May**, which might require investigation.

---

### ** Key Differences Between Matplotlib & Seaborn**
| Feature | Matplotlib | Seaborn |
|---------|-----------|---------|
| **Customization** | Manual (`grid=True, legend()`) | Built-in themes (`darkgrid`) |
| **Syntax Simplicity** | Requires more styling | Easier styling with fewer lines |
| **Works Without DataFrame?** |  Yes |  Yes |

This real-world example helps businesses **visualize sales trends** and **make data-driven decisions**! 

---

## ** Key Takeaways**
1. **Matplotlib requires manual styling**, while **Seaborn applies styles automatically** (grid, background, colors).
2. **Matplotlib needs explicit customization** (e.g., `edgecolors='black'` for scatter points).
3. **Seaborn is easier for working with DataFrames** (`data=df` argument).
4. **Seaborn has built-in themes** like `"darkgrid"`, `"whitegrid"`, `"ticks"`, etc.

Seaborn is great for quick, beautiful plots, whereas Matplotlib is best for full control over customization.

#### **2. Bar Plot: Product Sales Comparison**
Bar plots are great for comparing categories.

````python
# Data
data = {
    'Product': ['A', 'B', 'C'],
    'Sales': [100, 150, 80]
}
df = pd.DataFrame(data)

# Create the plot
sns.barplot(x='Product', y='Sales', data=df, palette='viridis')

# Add title and labels
plt.title('Product Sales Comparison')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
````

#### **3. Histogram: Exam Scores Distribution**
Histograms show the distribution of a single variable.

````python
# Data
scores = [55, 65, 75, 85, 75, 95, 85, 65, 75, 85, 95, 55]

# Create the plot
sns.histplot(scores, bins=5, kde=True, color='blue')

# Add title and labels
plt.title('Exam Scores Distribution')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.show()
````

#### **4. Scatter Plot: Height vs. Weight**
Scatter plots show relationships between two variables.

````python
# Data
data = {'Height': [150, 160, 170, 180, 190],
        'Weight': [50, 60, 70, 80, 90]}
df = pd.DataFrame(data)

# Create the plot
sns.scatterplot(x='Height', y='Weight', data=df, color='red')

# Add title and labels
plt.title('Height vs. Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
````

---

#### **5. Box Plot: Exam Scores by Class**
Box plots summarize data distribution and identify outliers.

````python
# Data
data = {'Class': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
        'Score': [85, 90, 78, 88, 84, 82, 79, 85, 88]}
df = pd.DataFrame(data)

# Create the plot
sns.boxplot(x='Class', y='Score', data=df, palette='pastel')

# Add title and labels
plt.title('Exam Scores by Class')
plt.xlabel('Class')
plt.ylabel('Score')
plt.show()
````


---

#### **6. Heatmap: Correlation Matrix**
Heatmaps show relationships between multiple variables.


````python
# Data
data = {'Math': [85, 78, 92, 88, 76],
        'Science': [89, 82, 95, 90, 80],
        'English': [78, 85, 88, 92, 84]}
df = pd.DataFrame(data)

# Calculate correlation matrix
correlation = df.corr()

# Create the plot
sns.heatmap(correlation, annot=True, cmap='coolwarm')

# Add title
plt.title('Correlation Matrix')
plt.show()
````
