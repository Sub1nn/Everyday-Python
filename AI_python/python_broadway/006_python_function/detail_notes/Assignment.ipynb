### Question 1: Using Functions and List (No Class)

**Objective:**
Write a Python program using **functions** and a list of subjects.

**Instructions:**

* Define a list of subjects:

  `["English", "Nepali", "Mathematics", "Science", "Social Studies", "E.P.H.", "Computer Science", "Optional Mathematics"]`

* Write a function `percentage_to_gpa(mark)` that converts percentage marks into GPA and grade according to the SEE grading scale.

* Write a function `input_marks(subjects)` that asks the user for marks for each subject and returns a list of marks.

* Calculate GPA for each subject, then calculate the average GPA.

* Print the GPA and grade for each subject, the average GPA, and the final grade.

| 35 to 39     | 1.6 | C     |
