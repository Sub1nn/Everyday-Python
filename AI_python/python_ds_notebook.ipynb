```python
from pathlib import Path
import json

try:
    BASE_DIR = Path(__file__).resolve().parent
except NameError:
    BASE_DIR = Path.cwd()

pDetails = BASE_DIR / 'person.json'
    

class Student :
    def __init__(self, name, age, address, skills, is_programmer, phone):
        self.name = name
        self.age = age
        self.address = address
        self.skills = skills
        self.is_programmer = is_programmer
        self.phone = phone

    def details(self):
        return {
            "name": self.name,
            "age": self.age,
            "address": self.address,
            "skills": self.skills,
            "is_programmer": self.is_programmer,
            "phone": self.phone
        }
    
def load_people(path: Path):
    if not path.exists() or path.read_text() == "":
        return []
    with open(path, 'r') as f:
        data = json.load(f)
    # return ([data] if isinstance(data, dict) else data if isinstance(data, list) else (_ for _ in ()).throw(ValueError("person.json must contain a dict or a list of dicts")))
    if isinstance(data, dict):
        return [data]
    if isinstance(data, list):
        return data
    raise ValueError("person.json must contain a dict or a list of dicts")
    
    
def rem_dups(records):
    seen = set()
    unique = []
    for r in records:
        phone = r.get("phone")
        if phone is None:
            unique.append(r)
            continue
        if phone not in seen:
            seen.add(phone)
            unique.append(r)
    return unique
    
data = load_people(pDetails)


s1 = Student('shyam', 25, 'Ktm', ['python', 'c++', 'ML', 'AI'], True, 9842456221)
s2 = Student("Hari", 18, "Pkr", ["AI", "C", "CS"], True, 9845623315)
s3 = Student("Subin", 35, "Finland", ["Python", "AIML"], True, +358445509013)

data.extend([s1.details(), s2.details(), s3.details()])
print(data)

data = rem_dups(data)

with open(pDetails, "w") as f:
    json.dump(data, f, indent=2)

print("saved to:", pDetails)
```

```python
s = "apples, pears # and bananas\ngrapes\nbananas !apples"

def strip_comments(s, markers):
    lines = s.split('\n')
    new = ""
    for line in lines:
        words = line.split()
        index = len(words)
        for i, w in enumerate(words):
            if any(w.startswith(m) for m in markers) or w in markers:
                index = i
                break
        new += "".join(words[:index]).rstrip() + "\n"
    print(lines)
    print(words)
    print(new)

strip_comments(s, ['#', '!'])
```

```python
with open ('file.txt', 'r') as f:
    text_file = f.read()
    words_array = text_file.split()
    words_count = {}
    for i in words_array:
        if i in words_count:
            words_count[i] += 1
        else:
            words_count[i] = 1
    for k, v in words_count.items():
        max_value = max(words_count.values())
        if v == max_value:
            print(k, v)
```
