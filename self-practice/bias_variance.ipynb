```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

# Generate synthetic data
np.random.seed(42)
X = np.linspace(-3, 3, 100)
y = 0.5 * X**2 + X + 2 + np.random.normal(0, 1, 100)

# Split data into training and test sets
X_train, X_test = X[:80], X[80:]
y_train, y_test = y[:80], y[80:]

# Underfitting: Linear model (high bias)
linear_model = LinearRegression()
linear_model.fit(X_train.reshape(-1, 1), y_train)
y_pred_linear = linear_model.predict(X_test.reshape(-1, 1))

# Overfitting: High-degree polynomial (degree=15, high variance)
poly_features = PolynomialFeatures(degree=15)
X_poly_train = poly_features.fit_transform(X_train.reshape(-1, 1))
X_poly_test = poly_features.transform(X_test.reshape(-1, 1))
poly_model = LinearRegression()
poly_model.fit(X_poly_train, y_train)
y_pred_poly = poly_model.predict(X_poly_test)

# Plot results
plt.scatter(X, y, color="blue", label="Data")
plt.plot(X_test, y_pred_linear, color="red", label="Underfitting (Linear)")
plt.plot(X_test, y_pred_poly, color="green", label="Overfitting (Degree=15)")
plt.legend()
plt.show()

# Evaluate performance
print("Linear Model MSE:", mean_squared_error(y_test, y_pred_linear))
print("Polynomial Model MSE:", mean_squared_error(y_test, y_pred_poly))

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

# Generate synthetic data
np.random.seed(42)
X = np.linspace(-3, 3, 100)
y = 0.5 * X**2 + X + 2 + np.random.normal(0, 1, 100)

# Split data into training and test sets
X_train, X_test = X[:80], X[80:]
y_train, y_test = y[:80], y[80:]

# Underfitting: Linear model (high bias)
linear_model = LinearRegression()
linear_model.fit(X_train.reshape(-1, 1), y_train)
y_pred_linear = linear_model.predict(X_test.reshape(-1, 1))

# Balanced Model: Quadratic (degree=2)
poly_features_balanced = PolynomialFeatures(degree=2)
X_poly_train_balanced = poly_features_balanced.fit_transform(X_train.reshape(-1, 1))
X_poly_test_balanced = poly_features_balanced.transform(X_test.reshape(-1, 1))
balanced_model = LinearRegression()
balanced_model.fit(X_poly_train_balanced, y_train)
y_pred_balanced = balanced_model.predict(X_poly_test_balanced)

# Plot results
plt.scatter(X, y, color="blue", label="Data")
plt.plot(X_test, y_pred_linear, color="red", label="Underfitting (Linear)")
plt.plot(X_test, y_pred_poly, color="green", label="Overfitting (Degree=15)")
plt.plot(X_test, y_pred_balanced, color="orange", label="Balanced (Degree=2)")
plt.legend()
plt.show()

# Evaluate performance
print("Linear Model MSE:", mean_squared_error(y_test, y_pred_linear))
print("Polynomial Model MSE:", mean_squared_error(y_test, y_pred_poly))
print("Balanced Model MSE:", mean_squared_error(y_test, y_pred_balanced))
```
