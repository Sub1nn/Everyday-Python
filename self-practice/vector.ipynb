```python
import numpy as np

def unit_vector(v):
    """Returns the unit vector of the given vector v."""
    magnitude = np.linalg.norm(v)  # Calculate the magnitude (norm)
    if magnitude == 0:
        raise ValueError("Zero vector has no unit vector")
    return v / magnitude  # Normalize the vector

# Example vector
v = np.array([3, 4, 0])  

# Compute unit vector
unit_v = unit_vector(v)

print("Original Vector:", v)
print("Unit Vector:", unit_v)
print("Magnitude of Unit Vector:", np.linalg.norm(unit_v))  # Should be 1
```

```python
import numpy as np
import matplotlib.pyplot as plt

def unit_vector(v):
    """Returns the unit vector of the given vector v."""
    magnitude = np.linalg.norm(v)
    if magnitude == 0:
        raise ValueError("Zero vector has no unit vector")
    return v / magnitude

# Example vector
v = np.array([3, 4])  
unit_v = unit_vector(v)

# Plotting
fig, ax = plt.subplots()
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label="Original Vector")
ax.quiver(0, 0, unit_v[0], unit_v[1], angles='xy', scale_units='xy', scale=1, color='b', label="Unit Vector")

# Formatting the graph
ax.set_xlim(-1, 4)
ax.set_ylim(-1, 4)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid(True, linestyle="--", linewidth=0.5)
ax.set_aspect('equal')
plt.legend()
plt.title("Unit Vector Visualization (2D)")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def unit_vector(v):
    """Returns the unit vector of the given vector v."""
    magnitude = np.linalg.norm(v)
    if magnitude == 0:
        raise ValueError("Zero vector has no unit vector")
    return v / magnitude

# Example vectors
a = np.array([3, 4])
b = np.array([1, 2])

# Normalize vector b
unit_b = unit_vector(b)

# Projection of a onto b
projection = np.dot(a, unit_b) * unit_b

# Plotting
fig, ax = plt.subplots()
ax.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label="Vector a")
ax.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='g', label="Vector b")
ax.quiver(0, 0, projection[0], projection[1], angles='xy', scale_units='xy', scale=1, color='b', label="Projection of a onto b")

# Formatting the graph
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 5)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid(True, linestyle="--", linewidth=0.5)
ax.set_aspect('equal')
plt.legend()
plt.title("Projection of Vector a onto Vector b")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# 3D Example vector
v3 = np.array([3, 4, 5])
unit_v3 = unit_vector(v3)

# 3D Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the vectors
ax.quiver(0, 0, 0, v3[0], v3[1], v3[2], color='r', label="Original Vector")
ax.quiver(0, 0, 0, unit_v3[0], unit_v3[1], unit_v3[2], color='b', label="Unit Vector")

# Formatting
ax.set_xlim([-1, 4])
ax.set_ylim([-1, 4])
ax.set_zlim([-1, 4])
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.legend()
plt.title("Unit Vector Visualization (3D)")
plt.show()
```
